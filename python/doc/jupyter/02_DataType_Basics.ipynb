# Tuple

1. Create a tuple

```python
t1 = (1,6,3,'hello',6,[2,3,4],(4,5,6),6,7)
print(t1) 
type(t1)
```

2. Tuple is iterable

```python
for i in t1:
    print(i, end=', ')
```

3. Tuple can be sliced
   t1[[start]:[stop]:[step]]
   by default: start=0; stop=end; step=1
   stop is excluded

```python
# slice a tuple
print(t1[5]) # individual element on index=5
print(t1[::]) # all three options are using default value: start=0, end=end, step=1
print(t1[4:6])  # tuple[start:stop:step], the stop element is excluded
print(t1[4:]) # all the way to the end
print(t1[::-1])
```

```python
n = 12345
s = str(n)
s[::-1],s[1::-1],s[3:0:-1]
```

```python
x = s[:2]+s[3:]
x
```

4. Tuple is immutable。

```python
t1[5] = "John" # TypeError: 'tuple' object does not support item assignment
```

5. there are some functions available for a tuple variable.

```python
print(t1)
print(t1.count(6))
print(t1.index("hello"))
```

6. tuple() built in function can be used to convert others to tuple.

```python
s = 'Hello, the world!'
print(tuple(s))
list1 = [1,2,3,4]
t1 = tuple(list1)
t1, type(t1)
```

```python
s2 = set()
print(type(s2))
print(len(s2))
s3 = set((1,2,3,4))
s3
```

```python
len(j1), j1.__len__()
```

```python
j = iter(j1) # j is iterator which is iterable
print(type(j))
print("next element:",next(j)) # if j is an iterator, you can use next() function to get next available item.
print("next element:",next(j))
print("next element:",next(j))
print("next element:",next(j))
print("next element:",next(j)) 
```

```python
print("next element:",next(j)) # you can not iterate iterator when it reaches the end of the iterator
```

```python
print("next element:",next(j,-1)) # where -1 is a default value for j
```

iterate: the processing to get each next avaible element<br>
iterator: is an python object which has bunch of element available for you to get.

```python
j=1
for i in j1: # where the i each individual element (this is also an iterate operation)
    i = i*2+123
    print(j, i) # one can use for-loop to iterate each element of an iterable object.
    j = j+1
```

```python
# set cannot be sliced
a = j1[3]
print(a)
```

```python
# set cannot have duplicated elements
s1 = {1,1,1,2,2,2,3,3,5,6,7}
print(s1)
```

```python

# set is mutable
s1.add(10)
print(s1)
s1.add(10)
print(s1)
s1.remove(3)
print(s1)
newSet = s1.pop()
print(s1, newSet)
newSet = s1.pop()
print(s1, newSet)

```
