# Comment
## single line comment
Single line comment can be used to avoid code from execution.

```python
a,b = 10,20
c= a+b
# c=a*b
print(f'{a} + {b} = {c}')
# print(f'{a} * {b} = {c}')
```

Can also be used as code documentation. from # to end of the line will not be considered as python executable code. 

```python
c = a + b # calculate c for a plus b
print(f'{a} + {b} = {c}')
```

# Variables
## Variable Naming Rules
1. start with alpha letter a-z,A-Z and _, no numeric letter allowed on first letter,
2. can be combination of a-z, A-Z, 0-9, and _
3. do not use python language reserved keyword as variable name.
4. avoid using built-in function name as variable name.
* [Python reserved keywords](https://realpython.com/lessons/reserved-keywords/)

## Why using variable?

Variables in Python are used to store values, which can be used and manipulated throughout a program. There are several reasons why you might want to use variables in Python:

1. To store values: Variables allow you to store values in memory, which can then be used and manipulated throughout your program.

2. To make code more readable: Using variables with descriptive names can make your code easier to read and understand. For example, instead of writing "x = 5" you could write "number_of_students = 5".

3. To make code more flexible: By using variables, you can easily change values throughout your code without having to manually update each occurrence. For example, if you have a variable that represents the sales tax rate, you can change the tax rate in one place and have it affect all calculations that use that variable.

4. To improve code organization: By grouping related values together in variables, you can make your code more organized and easier to maintain.

5. To pass values between functions: Variables can be used to pass values between functions, allowing you to write modular code that can be easily reused and tested.
